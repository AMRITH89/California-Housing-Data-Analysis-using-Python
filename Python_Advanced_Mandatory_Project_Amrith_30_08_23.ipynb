**Que 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
def median(data):
  median=0
  length=len(data)
  sorted_data=sorted(data)

  if length%2==0:
    n1=sorted_data[int(length/2)]
    n2=sorted_data[int((length/2)-1)]
    median=(n1+n2)/2
  else:
    median=sorted_data[length//2]
  return median

x=pd.read_excel("housing dataset.xlsx")
medianvalue=median(x.median_income)
print(medianvalue)
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
import numpy as np
import pandas as pd
hd=pd.read_excel("housing dataset.xlsx")

#"np.select(condlist, choicelist)"---For creating new column based on conditions
#filters-condlist
filters = [
   (hd.total_bedrooms<=10),
   (hd.total_bedrooms>=11) & (hd.total_bedrooms<1000),
   (hd.total_bedrooms>=1000)
]
#values-choicelist
values=["Small","Medium","Large"]

hd["total_bedroom_size"]=np.select(filters,values)  #np.select(condlist, choicelist)
print(hd)
```
